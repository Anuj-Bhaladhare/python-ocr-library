```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = [1, 2, 3]
y = [4, 5, 6]
```

```python
plt.plot(
    x, 
    y
    # color = "red", 
    # linestyle = "--"
)   # (1, 4)(2, 5)(3, 6)
plt.grid()
plt.show()
```

# Pyplot API

Univariate - Numerical

```python
data = {
    "Salary" : [25000, 30000, 37000, 28000, 39000, 48000, 55000, 52000, 35000, 26000, 27000, 31000, 33000, 29000, 40000, 36000, 34000, 45000, 47000, 50000]
}
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.shape  # Print size of the DataFrame.
```

# Line Plot

```python
plt.plot(
    df["Salary"], 
    color = "red", 
    marker = "o", 
    linestyle = ":", 
    linewidth = "2"
)

plt.grid()
plt.show()
```

# Histogram
